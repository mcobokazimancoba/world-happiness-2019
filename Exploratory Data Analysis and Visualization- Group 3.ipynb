## 3 🔷 PART C: RENAMING & SELECTION

### 📊 1. Bar Chart – Top 10 Happiest Countries

```python
# Ensure df_subset exists first
print(df_subset.head())

# Create sorted dataframe
happiness_sorted = df_subset.sort_values(by='happinessScore', ascending=False)

# Get top 10
top10 = happiness_sorted.head(10)

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x='happinessScore', y='country', data=top10)

plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")

plt.show()
```

### 📊 2. Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(df_subset['happinessScore'], bins=10)

plt.title("Distribution of Happiness Score")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```

### 📝 Answer:
### The distribution appears slightly skewed / approximately normal

### 📊 3. Box Plot

```python
top_regions = (
    df_cleaned.groupby('region')['happinessScore']
    .mean()
    .sort_values(ascending=False)
    .head(20)  # limit to top 20 for readability
    .index
)

df_plot = df_cleaned[df_cleaned['region'].isin(top_regions)]

plt.figure(figsize=(14, 6))
sns.boxplot(x='region', y='happinessScore', data=df_plot, order=top_regions)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.title("Happiness Score by Region (Top 20)", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 📝 Answers:
### Region with highest median: Finland
### Outliers:  No

```python
df_cleaned.groupby('region')['happinessScore'].median().idxmax()
```
